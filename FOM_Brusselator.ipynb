```python
from fenics import *
import matplotlib.pyplot as plt

# 1. Create a Mesh (The 'Order' of your model)
# 50 points across the interval [0, 1]
mesh = UnitIntervalMesh(50)

# 2. Define Function Space (Linear Lagrange elements)
V = FunctionSpace(mesh, 'P', 1)

# 3. Define Boundary Conditions (u=0 at both ends)
bc = DirichletBC(V, Constant(0), 'on_boundary')

# 4. Define the Variational Problem (The "Weak Form")
# Trial function (u) and Test function (v)
u = TrialFunction(V)
v = TestFunction(V)
f = Expression("sin(pi*x[0])", degree=2) # A dummy source term
nu = Constant(0.1) # Diffusion coefficient

# The "Bilinear Form" a(u, v) and "Linear Functional" L(v)
# Equivalent to: Integral( nu * grad(u) * grad(v) ) dx
a = nu * dot(grad(u), grad(v)) * dx
L = f * v * dx

# 5. Assemble the "Full Order" Matrix and Vector
A = assemble(a) # This is a 50x50 Sparse Matrix
b = assemble(L) # This is a vector of length 50
bc.apply(A, b)  # Apply the boundary conditions to the system

# 6. Solve for u
u_sol = Function(V)
solve(A, u_sol.vector(), b)

# 7. Plot the result
plot(u_sol)
plt.title("Dummy FOM Solution (Stationary Step)")
plt.show()
```
